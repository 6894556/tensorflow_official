## vectorization

wind volumn, wind direction으로 바람 vct

```python
wv = df2.pop("wv (m/s)")
max_wv = df2.pop("max. wv (m/s)")
wd = df2.pop("wd (deg)")
```

```python
# to radian
wd_rad = wd * np.pi / 180
```

```python
# W
df2["Wx"] = wv * np.cos(wd_rad)
df2["Wy"] = wv * np.sin(wd_rad)
```

```python
# max W
df2["max Wx"] = max_wv * np.cos(wd_rad)
df2["max Wy"] = max_wv * np.sin(wd_rad)
```

```python
# dt변환 format 확인
pd.to_datetime(df2["Date Time"], format="%d.%m.%Y %H:%M:%S")
```

```python
# Date Time열 dt화 & df2에서 pop
date_time = pd.to_datetime(df2.pop("Date Time"), format="%d.%m.%Y %H:%M:%S")
```

```python
# sec로 변환
timestamp_s = date_time.map(pd.Timestamp.timestamp)
```

```python
day_s = 24*60*60
year_s = (365.2425)*day_s
```

```python
df2["Day sin"] = np.sin(timestamp_s * (2 * np.pi / day_s))
df2["Day cos"] = np.cos(timestamp_s * (2 * np.pi / day_s))
```

```python
df2["Year sin"] = np.sin(timestamp_s * (2 * np.pi / year_s))
df2["Year cos"] = np.cos(timestamp_s * (2 * np.pi / year_s))
```
